```python
import numpy
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# simple experiment - doubling

z = 0.0 + 0.0j
c = 0.1 + 0.75j

# add initial value to list
values = [z]

for i in range(100):
    z = z*z + c
    #z = numpy.sqrt(z)
    #z = z / 2

    values.append(z)
    pass
```

```python
# show values 

#values
values = numpy.array(values)
```

```python
# plot values
```

```python
plt.figure(figsize=(10,10))
s = 10
plt.xlim(-s,s)
plt.ylim(-s,s)

plt.plot(values.real, values.imag, '-ok', alpha=0.2)
```
